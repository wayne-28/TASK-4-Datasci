```python
import pandas as pd
import matplotlib.pyplot as plt
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk.tokenize import word_tokenize
import nltk
nltk.download('punkt')
nltk.download('vader_lexicon')

# Sample social media data
data = {
    'Date': ['2024-01-01', '2024-01-02', '2024-01-03', '2024-01-04', '2024-01-05'],
    'Text': [
        "I love the new iPhone, it's amazing!",
        "The customer service was terrible.",
        "Great experience with the brand!",
        "Not satisfied with the product quality.",
        "Excited to try out the new features."
    ]
}

df = pd.DataFrame(data)

# Sentiment analysis
sid = SentimentIntensityAnalyzer()
df['Sentiment'] = df['Text'].apply(lambda x: sid.polarity_scores(x)['compound'])

# Visualization
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Sentiment'], marker='o')
plt.xlabel('Date')
plt.ylabel('Sentiment Score')
plt.title('Sentiment Analysis')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show()
```
